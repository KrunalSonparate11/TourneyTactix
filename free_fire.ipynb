```python
# Replace non-numeric values in the Age column with NaN
character_data['Age'] = pd.to_numeric(character_data['Age'], errors='coerce')

# Drop rows where Age is NaN for plotting a clean age distribution
cleaned_character_data = character_data.dropna(subset=['Age'])
```

```python
# Plot age distribution (cleaned data)
plt.figure(figsize=(10, 5))
# Histogram to show the distribution of character ages, with KDE to indicate the density curve.
sns.histplot(cleaned_character_data['Age'].astype(int), bins=15, kde=True, color='skyblue')
plt.title('Age Distribution of Characters')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Part 1: Analyze character data for tournament best fits
# Filter characters with unique, impactful survival abilities at higher levels
character_data['Level 8 Effectiveness'] = character_data['Level 8'].str.extract(r'(\d+)%').astype(float)

# Sort characters by the effectiveness of their level 8 abilities
top_characters = character_data.sort_values(by='Level 8 Effectiveness', ascending=False)

# Display the top 5 characters based on their level 8 ability effectiveness
print("Top 5 Characters for Tournaments:")
print(top_characters[['Character Name', 'Survival Ability', 'Level 8 Effectiveness']].head(5))

# Part 2: Analyze weapon data for tournament best fits
# Calculate a performance score for each weapon (weighted average of DAMAGE, ACCURACY, and RATE OF FIRE)
weapons_data['Performance Score'] = (weapons_data['DAMAGE'] * 0.5 +
                                     weapons_data['ACCURACY'] * 0.3 +
                                     weapons_data['RATE_OF_FIRE'] * 0.2)

# Sort weapons by the performance score
top_weapons = weapons_data.sort_values(by='Performance Score', ascending=False)

# Display the top 5 weapons based on the performance score
print("\nTop 5 Weapons for Tournaments:")
print(top_weapons[['Weapon_Name', 'Weapon_type', 'DAMAGE', 'ACCURACY', 'RATE_OF_FIRE', 'Performance Score']].head(5))
```

```python
# Define character roles based on survival abilities
def categorize_ability(ability):
    if 'increase' in ability.lower() or 'penetration' in ability.lower():
        return 'Offensive'
    elif 'reduce' in ability.lower() or 'armor' in ability.lower():
        return 'Defensive'
    elif 'heal' in ability.lower() or 'HP' in ability.lower():
        return 'Support'
    else:
        return 'Utility'

# Apply categorization logic to each character
character_data['Ability Type'] = character_data['Survival Ability'].apply(categorize_ability)

# Sort characters by effectiveness at level 8
character_data['Level 8 Effectiveness'] = character_data['Level 8'].str.extract(r'(\d+)%').astype(float)
top_characters = character_data.sort_values(by='Level 8 Effectiveness', ascending=False)

# Select top characters by type (up to 2 per category)
selected_characters = pd.concat([
    top_characters[top_characters['Ability Type'] == 'Offensive'].head(2),
    top_characters[top_characters['Ability Type'] == 'Defensive'].head(1),
    top_characters[top_characters['Ability Type'] == 'Support'].head(1)
])

# Display the selected characters for tournament use
print("Selected Characters for Tournament (Optimal Skill Combination):")
print(selected_characters[['Character Name', 'Survival Ability', 'Ability Type', 'Level 8 Effectiveness']])
```

```python
from itertools import combinations

# Categorize abilities as done previously
def categorize_ability(ability):
    if 'increase' in ability.lower() or 'penetration' in ability.lower():
        return 'Offensive'
    elif 'reduce' in ability.lower() or 'armor' in ability.lower():
        return 'Defensive'
    elif 'heal' in ability.lower() or 'HP' in ability.lower():
        return 'Support'
    else:
        return 'Utility'

# Apply the categorization
character_data['Ability Type'] = character_data['Survival Ability'].apply(categorize_ability)

# Extract relevant columns and sort by Level 8 Effectiveness
character_data['Level 8 Effectiveness'] = character_data['Level 8'].str.extract(r'(\d+)%').astype(float)
character_data = character_data.sort_values(by='Level 8 Effectiveness', ascending=False)

# Generate all combinations of 4 characters at a time
character_combinations = list(combinations(character_data.index, 4))

# Create a DataFrame to store the selected character combinations and their combined effectiveness
combination_results = []

for combo in character_combinations:
    selected_combo = character_data.loc[list(combo)]
    effectiveness_sum = selected_combo['Level 8 Effectiveness'].sum()
    combination_results.append({
        'Characters': selected_combo['Character Name'].tolist(),
        'Total Effectiveness': effectiveness_sum,
        'Details': selected_combo[['Character Name', 'Survival Ability', 'Ability Type', 'Level 8 Effectiveness']]
    })

# Sort the combinations by total effectiveness
top_combinations = sorted(combination_results, key=lambda x: x['Total Effectiveness'], reverse=True)[:5]

# Display the top 4-character combinations
print("Top 4-Character Combinations for Tournaments:")
for i, combo in enumerate(top_combinations):
    print(f"\nCombination {i+1}:")
    print(f"Total Effectiveness: {combo['Total Effectiveness']:.2f}")
    print(combo['Details'])
```
